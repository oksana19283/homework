### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
from collections import Counter
def maximum(arr):
    max=arr[0]
    for i in range(1,len(arr)):
        if arr[i]>max:
            max=arr[i]
    return max
def minimum(arr):
    min=arr[0]
    for i in range(1,len(arr)):
        if arr[i]<min:
            min=arr[i]   
    return min

def averege(arr):
    s=0
    for i in arr:
        s=s+i
    av=s/len(arr)
    return av
def mediana(arr):
    arr.sort()
    if len(arr)%2==0:
        i=int(len(arr)/2)
        med=(arr[i]+arr[i-1])/2
    return med
def frequency_number(arr):
    counter = Counter(arr)
    return counter
def standard_deviation(arr,a):
    s=0
    for i in arr:
        m=(i-a)**2
        s=s+m
    s=(s/len(arr))**0.5
    return s

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,10) for i in range(1000)]

a=averege(list_with_random_numbers)
print(list_with_random_numbers)
print('maximum=',maximum(list_with_random_numbers))
print('minimum=',minimum(list_with_random_numbers))
print('avarage=',a)
print('mediana=',mediana(list_with_random_numbers))
print('frequency_number=',frequency_number(list_with_random_numbers))  
print('standard_deviation=',standard_deviation(list_with_random_numbers,a))
```
