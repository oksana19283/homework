```python
today = datetime.date.today()
for i in range(0,12):
    date = today+ relativedelta(months=-1)
    date_d=today+ relativedelta(months=-1)
    s=0
    i=0
    
    while date < today:
        a=date.strftime("%Y%m%d")
        r=requests.get(f'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?date={a}&json).json()').json()
        for el in r:
            if el['cc']=='USD':
                s=s+el['rate']
                i=i+1
                
        date += datetime.timedelta(days=1)  
        
    s=s/i
    
    
    while date_d < today:
        a=date_d.strftime("%Y%m%d")
        r=requests.get(f'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?date={a}&json).json()').json()
                
        for el in r:
            if el['cc']=='USD':
                deviation=(el['rate']-s)**2
                sum_dev=sum_dev+deviation
        date_d += datetime.timedelta(days=1)  

    sum_dev=(sum_dev/i)**0.5
    
    
    print('average value',today+ relativedelta(months=-1),'-',today,'=',s)
    print('course deviation',today+ relativedelta(months=-1),'-',today,'=',sum_dev)
    
    today=today+ relativedelta(months=-1)
```
